```python
# Question 2
# Plot a line plot showing the sales trend in company 1 and 2:
# days = [1,2,3,4,5,6,7] #days of d week
# sales_1 = [160,150,140,145,175,165,180] #sales of company1
# sales_2 = [70,90,160,150,140,145,175]  #sales of company2
```

```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
```

```python
plt.title("Sale_1 vs Sales_2")
plt.xlabel("Days")
plt.ylabel("Sales Volume")
plt.plot(days, sales_1, 'ro-')
plt.plot(days, sales_2, 'bo-')
```

```python
# Question 3
# Create a 3 by 3 subplots:
# multiple plots
# x = [1,2,3,4]
# y1 = [4,3,2,1]
# y2 = [10,20,30,40]
# y3 = [40,30,20,10]
# y4 = [1,2,1,2]
# y5 = [40,70,90,70]
```

```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
```

```python
plt.subplot(3, 3, 1)
plt.plot(x, y1, 'r')
plt.subplot(3, 3, 2)
plt.plot(x, y2, 'r+--')
plt.subplot(3, 3, 3)
plt.plot(x, y3, 'r*-')
plt.subplot(3, 3, 4)
plt.plot(x, y4, 'g')
plt.subplot(3, 3, 5)
plt.plot(x, y5, 'g+--')
plt.subplot(3, 3, 6)
plt.plot(x, y1, 'g*-')
plt.subplot(3, 3, 7)
plt.plot(x, y2, 'b')
plt.subplot(3, 3, 8)
plt.plot(x, y3, 'b+--')
plt.subplot(3, 3, 9)
plt.plot(x, y4, 'b*-')
```
