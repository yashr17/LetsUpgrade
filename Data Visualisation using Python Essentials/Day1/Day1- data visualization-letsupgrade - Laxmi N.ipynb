# Line Plot


*Displays information as a series of datapoints called markers, connected by straight line segments

```python
## Simple Examples

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.plot(x,y,'go--', linewidth=1, markersize=5)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine waveform") 

# Plot the points using matplotlib 
plt.plot(x, y, color='r') 
plt.show()
```

```python
x=np.arange(0,10)
y= 3*x + 10
## plt plot
plt.plot(x,y, 'ro--')
#plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X-axis', color='blue')
plt.ylabel('Y-axis', color='blue')
plt.title('2d-Diagram', color='blue')
```

```python
## Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```
